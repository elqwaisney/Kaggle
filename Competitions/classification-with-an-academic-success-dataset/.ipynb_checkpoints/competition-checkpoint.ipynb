## Feature Engineering

Feature engineering is a critical step in preparing your data for machine learning models. Here are some general strategies and specific suggestions based on the dataset context you provided:

### General Strategies:
1. **Handling Missing Values**: Determine if any columns have missing values and decide how to handle them (e.g., imputation, removal).
2. **Encoding Categorical Variables**: Convert categorical variables into numerical format using techniques like one-hot encoding or label encoding.
3. **Scaling and Normalization**: Normalize or standardize numerical features to ensure they contribute equally to the model.
4. **Creating New Features**: Derive new features from existing ones to capture additional information (e.g., interaction terms, polynomial features).
5. **Feature Selection**: Choose the most relevant features using methods like correlation analysis, feature importance from models, or dimensionality reduction techniques like PCA.

### Specific Suggestions for Your Dataset:

1. **Handling Categorical Variables**:
   - One-hot encode variables like 'Marital status', 'Application mode', 'Nacionality', 'Mother's qualification', 'Father's qualification', 'Mother's occupation', 'Father's occupation', 'Gender'.
   - For 'Target', convert it into numerical labels if not already done (e.g., Dropout = 0, Enrolled = 1, Graduate = 2).

2. **Feature Transformation**:
   - Log-transform skewed features like 'Admission grade' if they have a long tail distribution.
   - Binning numerical variables like 'Age at enrollment' into age groups.

3. **Interaction Features**:
   - Create interaction terms between 'Mother's qualification' and 'Father's qualification' to capture combined parental influence.
   - Interaction between 'Admission grade' and 'Previous qualification (grade)' to see how prior performance affects current performance.

4. **Aggregating Features**:
   - Calculate total credits for each student by summing 'Curricular units 1st sem (credited)' and 'Curricular units 2nd sem (credited)'.
   - Create a ratio of 'Curricular units 1st sem (credited)' to 'Curricular units 1st sem (enrolled)' to capture course completion rates.

5. **Handling `Curricular units 1st sem (without evaluations)`**:
   - Create a binary feature indicating whether a student had any courses without evaluations.
   - Calculate the proportion of courses without evaluations to total enrolled courses.

6. **Temporal Features**:
   - If you have the enrollment year, calculate the number of years since enrollment or use it to capture trends over time.

### Example Code:
Here’s some example code implementing some of these feature engineering steps:

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Example DataFrame
df = train.copy()

# One-hot encoding categorical variables
categorical_features = ['Marital status', 'Application mode', 'Nacionality', 'Mother\'s qualification', 'Father\'s qualification', 'Mother\'s occupation', 'Father\'s occupation', 'Gender']
one_hot = OneHotEncoder()

# Log-transforming Admission grade
df['Log Admission grade'] = np.log1p(df['Admission grade'])

# Creating interaction terms
df['Parent Qualification Interaction'] = df['Mother\'s qualification'] * df['Father\'s qualification']

# Aggregating features
df['Total Credits'] = df['Curricular units 1st sem (credited)'] + df['Curricular units 2nd sem (credited)']
df['Completion Rate 1st Sem'] = df['Curricular units 1st sem (credited)'] / df['Curricular units 1st sem (enrolled)']
df['Completion Rate 2nd Sem'] = df['Curricular units 2nd sem (credited)'] / df['Curricular units 2nd sem (enrolled)']

# Binary feature for courses without evaluations
df['Has Courses Without Evaluations'] = df['Curricular units 1st sem (without evaluations)'] > 0

# Creating a ColumnTransformer for pre-processing
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['Log Admission grade', 'Total Credits', 'Completion Rate 1st Sem', 'Completion Rate 2nd Sem']),
        ('cat', one_hot, categorical_features)
    ]
)

# Apply transformations
df_transformed = preprocessor.fit_transform(df)

# Convert to DataFrame for easier interpretation
df_transformed = pd.DataFrame(df_transformed, columns=preprocessor.get_feature_names_out())

print(df_transformed.head())
```

````python
import pandas as pd

# Function to combine columns and drop redundant ones
def combine_and_drop(df, col1, col2, new_col):
    df[new_col] = df[col1] + df[col2]
    df.drop(columns=[col1, col2], inplace=True)
    return df

# Columns to combine and drop
columns_to_combine = [
    ('Curricular units 1st sem (grade)', 'Curricular units 2nd sem (grade)', '1st year grades'),
    ('Curricular units 1st sem (without evaluations)', 'Curricular units 2nd sem (without evaluations)', '1st year courses (without evaluations)'),
    ('Curricular units 1st sem (evaluations)', 'Curricular units 2nd sem (evaluations)', '1st year courses (evaluations)'),
    ('Curricular units 1st sem (enrolled)', 'Curricular units 2nd sem (enrolled)', '1st year courses (enrolled)'),
    ('Curricular units 1st sem (credited)', 'Curricular units 2nd sem (credited)', '1st year courses (credited)'),
    ('Curricular units 1st sem (approved)', 'Curricular units 2nd sem (approved)', '1st year courses (approved)'),
    ('Previous qualification (grade)', 'Admission grade', 'grades before admission')
]

# Apply transformations on train dataset
for col1, col2, new_col in columns_to_combine:
    train = combine_and_drop(train, col1, col2, new_col)

# Apply transformations on test dataset
for col1, col2, new_col in columns_to_combine:
    test = combine_and_drop(test, col1, col2, new_col)

# Display the first few rows of train and test datasets to verify
train.head()
test.head()
````

````python
#Make an interaction term between app mode and app order.
train['App_modXord'] = train['Application mode']* train['Application order']
train.drop(columns= ['Application mode', 'Application order'], inplace = True)

test['App_modXord'] = test['Application mode']* test['Application order']
test.drop(columns= ['Application mode', 'Application order'], inplace = True)
````

````python
# Define the columns and interaction terms
columns_to_combine = [
    ('Mother\'s qualification', 'Father\'s qualification', 'Parents\'qualifications'),
    ('Mother\'s occupation', 'Father\'s occupation', 'Parents\'occupation')
]

# Apply transformations on train dataset
for col1, col2, new_col in columns_to_combine:
    train[new_col] = train[col1] * train[col2]
    train.drop(columns=[col1, col2], inplace=True)

# Apply transformations on test dataset
for col1, col2, new_col in columns_to_combine:
    test[new_col] = test[col1] * test[col2]
    test.drop(columns=[col1, col2], inplace=True)
````

````python

# Define a function for binning age
def bin_age(df, bin_edges, bin_labels, col_name):
    df[col_name + '_bin'] = pd.cut(df[col_name], bins=bin_edges, labels=bin_labels, right=False)
    return df

# Define bin edges and labels
bin_edges = [17, 20, 25, 30, 35, 40, 50, 70]
bin_labels = ['17-20', '21-25', '26-30', '31-35', '36-40', '41-50', '51-70']

# Apply binning to train dataset
train = bin_age(train, bin_edges, bin_labels, 'Age at enrollment')

# Apply binning to test dataset
test = bin_age(test, bin_edges, bin_labels, 'Age at enrollment')
````
